```python
#make a 5 second timer that ends in BOOM!

import time

print(5)
time.sleep(1)
print(4)
time.sleep(1)
print(3)
time.sleep(1)
print(2)
time.sleep(1)
print(1)
time.sleep(1)
print("BOOM YOU DIE")
```

```python
#use a for in range loop to make a 60 second timer that ends in boom!
import time

num = 60
speed = 0.5
for i in range(num):
    time.sleep(speed)
    print(num - i)

print("boom")
```

```python
#use input to enter how much time you want to wait like a stopwatch

import time
num = input("insert seconds")
speed = 0.5
for i in range(int(num)):
    time.sleep(speed)
    print(int(num) - i)

print("boom")
```
